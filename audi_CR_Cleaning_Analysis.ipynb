```python
import pandas as pd

def clean_data(df):
    # Drop column: 'Main_CR_Number'
    df = df.drop(columns=['Main_CR_Number'])
    # Replace missing values with 0 in column: 'Capital'
    df = df.fillna({'Capital': 0})
    # Remove leading and trailing whitespace in columns: 'CR_Type', 'Commercial_Name' and 4 other columns
    df['CR_Type'] = df['CR_Type'].str.strip()
    df['Commercial_Name'] = df['Commercial_Name'].str.strip()
    df['CR_Legal_Structure_AR'] = df['CR_Legal_Structure_AR'].str.strip()
    df['CR_Region_Name_AR'] = df['CR_Region_Name_AR'].str.strip()
    df['CR_City_Name_AR'] = df['CR_City_Name_AR'].str.strip()
    df['CR_Issue_Date_GR'] = df['CR_Issue_Date_GR'].str.strip()
    # Replace missing values with "Unknown" in column: 'CR_Legal_Structure_AR'
    df = df.fillna({'CR_Legal_Structure_AR': "Unknown"})
    # Drop rows with missing data in column: 'NationalNo'
    df = df.dropna(subset=['NationalNo'])
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    return df

# Loaded variable 'df' from URI: /Users/muhannadalshammari/Desktop/Commercial_registration_2023_Q1.csv
df = pd.read_csv(r'/Users/muhannadalshammari/Desktop/Commercial_registration_2023_Q1.csv')

df_clean = clean_data(df.copy())
df_clean.head()


df.to_csv('cleaned_data.csv', index=False)
```
